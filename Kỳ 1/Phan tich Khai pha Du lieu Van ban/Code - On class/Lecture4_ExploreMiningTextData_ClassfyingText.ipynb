```python
dictionary = ['tôi','thích', 'món', 'phở', 'việt', 'nam', 'là', 'ăn','ngon', 'của']
l1 = [str.lower(i) for i in "Tôi thích món phở Việt Nam".split()]
l2 = [str.lower(i) for i in "Phở là món ăn ngon của Việt Nam".split()]

output1 = []
output2 = []
for i in dictionary:

    output1.append(int(i in l1))
    output2.append(int(i in l2))
print(output1)
print(output2)
```
