Áp dụng SVD trong LSI để tóm tắt chủ đề của tập văn bản bằng cách ánh xạ tài liệu vào không gian k chiều (chủ đề).

Ví dụ: Viết code Python để biểu diễn 15 tài liệu sau trong không gian 2 chiều và 3 chiều bằng SVD (k=2 và k=3). Sau đó, phân tích các từ quan trọng nhất trong từng chiều để gợi ý “chủ đề” tiềm ẩn.

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD

documents = ['information retrieval system',
'retrieval of information',
'system design',
'design of database system',
'information theory and retrieval',
'machine learning and information retrieval',
'neural networks for text classification',
'deep learning for natural language processing',
'text mining and knowledge discovery',
'database design and normalization',
'retrieval models in information retrieval',
'semantic search and latent semantic indexing',
'language models for information retrieval',
'system architecture and distributed systems',
'clustering of documents using machine learning']

# 1. Biểu diễn tài liệu dưới dạng ma trận Term-Document sử dụng TF-IDF
# TF-IDF (Tần số thuật ngữ-Tần số nghịch đảo tài liệu) là một kỹ thuật đánh trọng số cho các từ.
# Nó giúp làm nổi bật các từ quan trọng trong tài liệu bằng cách giảm trọng số của các từ phổ biến.
print("=== 1. Tạo ma trận Term-Document bằng TF-IDF ===")
vectorizer = TfidfVectorizer(stop_words=['and', 'of', 'in', 'for', 'the'])
X = vectorizer.fit_transform(documents)

# Lấy tên của các từ (features) sau khi đã xử lý
feature_names = vectorizer.get_feature_names_out()
print(f"Kích thước ma trận ban đầu: {X.shape[0]} tài liệu x {X.shape[1]} từ")
print("\n" + "="*50 + "\n")
```

```python
print(X)
```

```python
# 2. Phân tích SVD với k = 2 (Không gian 2D)
# Truncated SVD là một biến thể của SVD, được sử dụng để giảm chiều ma trận
# Nó sẽ chuyển ma trận Term-Document thành một ma trận mới có 2 chiều
# nơi mỗi chiều đại diện cho một "chủ đề" tiềm ẩn.
print("=== 2. Phân tích SVD với k = 2 (Không gian 2D) ===")
svd_2d = TruncatedSVD(n_components=2)
# Phép toán fit_transform sẽ giảm chiều ma trận X
document_vectors_2d = svd_2d.fit_transform(X)

print("Biểu diễn tài liệu trong không gian 2D:")
df_2d = pd.DataFrame(document_vectors_2d, columns=['Chiều 1', 'Chiều 2'], index=documents)
print(df_2d)
print("\n")

# Phân tích các chiều (chủ đề) tiềm ẩn
print("Phân tích các chủ đề tiềm ẩn (k = 2):")
for i, comp in enumerate(svd_2d.components_):
    # Lấy các từ quan trọng nhất trong mỗi chiều
    # comp là một mảng 1D, argsort() trả về chỉ số của các giá trị được sắp xếp
    top_words_idx = comp.argsort()[-5:][::-1]
    top_words = [feature_names[j] for j in top_words_idx]
    
    # Gợi ý chủ đề dựa trên các từ khóa
    theme = ""
    if "information" in top_words and "retrieval" in top_words:
        theme = "Chủ đề về 'Truy hồi thông tin'"
    elif "system" in top_words and "design" in top_words:
        theme = "Chủ đề về 'Thiết kế hệ thống'"
    else:
        theme = "Chủ đề hỗn hợp hoặc không rõ ràng"

    print(f"  - Chiều {i+1} (Chủ đề tiềm ẩn): {top_words}")
    print(f"    Gợi ý: {theme}")

print("\n" + "="*50 + "\n")
```

```python
# 3. Phân tích SVD với k = 3 (Không gian 3D)
# Tương tự, ta sẽ giảm chiều ma trận xuống 3 chiều để có một biểu diễn chi tiết hơn
print("=== 3. Phân tích SVD với k = 3 (Không gian 3D) ===")
svd_3d = TruncatedSVD(n_components=3)
# Giảm chiều ma trận X xuống 3 chiều
document_vectors_3d = svd_3d.fit_transform(X)

print("Biểu diễn tài liệu trong không gian 3D:")
df_3d = pd.DataFrame(document_vectors_3d, columns=['Chiều 1', 'Chiều 2', 'Chiều 3'], index=documents)
print(df_3d)
print("\n")

# Phân tích các chiều (chủ đề) tiềm ẩn
print("Phân tích các chủ đề tiềm ẩn (k = 3):")
for i, comp in enumerate(svd_3d.components_):
    # Lấy các từ quan trọng nhất trong mỗi chiều
    top_words_idx = comp.argsort()[-5:][::-1]
    top_words = [feature_names[j] for j in top_words_idx]

    # Gợi ý chủ đề dựa trên các từ khóa
    theme = ""
    if "information" in top_words and "retrieval" in top_words:
        theme = "Chủ đề về 'Truy hồi thông tin'"
    elif "system" in top_words and "design" in top_words:
        theme = "Chủ đề về 'Thiết kế và phát triển hệ thống'"
    elif "learning" in top_words and "machine" in top_words:
        theme = "Chủ đề về 'Học máy và phân tích dữ liệu'"
    else:
        theme = "Chủ đề hỗn hợp hoặc không rõ ràng"

    print(f"  - Chiều {i+1} (Chủ đề tiềm ẩn): {top_words}")
    print(f"    Gợi ý: {theme}")
```
